# URL Document Loader

LangChain은 웹에 기록된 글도 텍스트 형식으로 가져와서 LLM에 활용할 수 있습니다.

```python
from langchain.document_loaders import WebBaseLoader

loader = WebBaseLoader('https://www.nocutnews.co.kr/news/6087302')
data = loader.load()

print(data[0].page_content)
```

```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader('/content/[정책브리프 2021-04] 탄소중립 대응을 위한 정부 정책과 동향 (1).pdf')
pages = loader.load_and_split()

print('Total pages: ' + str(len(pages)))
print('Page content: ' + pages[10].page_content)
print(pages[10].metadata)
```

```python
# CSV Loader

from langchain.document_loaders import CSVLoader
loader = CSVLoader(file_path='/content/titanic3.csv',
                   csv_args={
                       'delimiter':',',
                       'quotechar':'"',
                       'fieldnames': ['pclass','survived','name','sex','age','sibsp','parch','ticket','fare','cabin','embarked','boat','body','home.dest']
})

data = loader.load()
```
