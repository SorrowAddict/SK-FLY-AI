## 스타일 지정하는 법
fmt = “[color][marker][line]”

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + 0.5)
y3 = np.sin(x + 1)
plt.plot(x, y1, 'r--', label='Sin A')
plt.plot(x, y2, 'b.', label='Sin B')
plt.plot(x, y3, 'g-.', label='Sin C')
plt.title('sin 함수')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.legend(loc='upper right')       # 범례 위치 변경
plt.show()
```

# Figurer과 Axes
## matplotlib 객체
‣ Figurer 객체, Axes 객체, Axis 객체 등으로 구성\
‣ Figure 객체는 그림이 그려지는 전체 영역을 말하고, Axes는 하나의 그래프를 의미.\
‣ Axis는 그래프를 구성하는 x, y축을 의미.\
‣ Figure 객체는 한 개 이상의 Axes 객체를 포함.

```python
x = np.arange(1, 30)
y = np.log(x)
plt.plot(x, y)
fig = plt.gcf()
fig.set_size_inches(8,3)    #그래프 크기 설정
plt.show()
```

# 여러 개의 그래프 그리기
subplot(row, column, index)를 호출하면 row x column 개수의 그래프를 그려서 index의 위치에 그래프를 추가

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
plt.style.use('ggplot')
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)        #subplot(행, 열, 인덱스)
plt.plot(x, y, 'r-')
plt.subplot(2, 2, 2)
plt.plot(x, z, 'g--')
plt.subplot(2, 2, 3)
plt.plot(x, z, 'b-.')
plt.show()
```

# 막대그래프
bar(), barh()

```python
x = ['A','B','C','D']
y = [5. , 25. , 50. , 20.]
plt.bar(x, y)
plt.title('막대그래프')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.subplot(1, 2, 1)        #subplot(행, 열, 인덱스)

name = ['A','B','C','D']
y = [5. , 25. , 50. , 20.]
z = [8. , 20. , 25. , 6.]
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(12, 5)
ax[0].bar(name, y, color='#33aaff')
ax[1].bar(name, z, color=['r', 'g', 'b', 'k'])
plt.subplot(1, 2, 2)        #subplot(행, 열, 인덱스)

plt.show()
```

# 박스 플롯
boxplot()

###• 색상
• 박스 내부의 색상을 설정할 수 있음\
• patch_artist : 박스 내부를 색상으로 채운다.\
• sym : 데이터를 표시하는 색상과 모양을 문자열로 설정.

```python
x = np.random.normal(10, 3, 100)
y = np.random.normal(13, 3, 100)
z = np.random.normal(8, 5, 100)
plt.boxplot([x, y, z], patch_artist=True, sym='ro')
plt.xticks([1, 2, 3], ['A', 'B', 'C'])
plt.show()
```
