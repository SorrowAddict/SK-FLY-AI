```python
from numpy import dot
from numpy.linalg import norm
import numpy as np

def cos_sim(A,B):
  return dot(A,B)/(norm(A)*norm(B))
```

## model들 뭉쳐있는 github 같은 꿀 같은 곳 !!
### https://huggingface.co/

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings

model_name = 'BAAI/bge-small-en'                # https://huggingface.co/BAAI/bge-small-en 에서 가져다 쓸 모델 이름
model_kwargs = {'device':'cpu'}
encode_kwargs = {'normalize_embeddings':True}

hf = HuggingFaceBgeEmbeddings(
    model_name = model_name,
    model_kwargs = model_kwargs,
    encode_kwargs = encode_kwargs
)
```

```python
embeddings = hf.embed_documents(
              [
                  'today is monday',
                  'weather is nice today',
                  "what's the problem",
                  'langchain is useful',
                  'Hello World',
                  'my name is morris'
              ]
)
```

```python
BGE_query_q = hf.embed_query('Hello? who is this?')
BGE_query_a = hf.embed_query('hi this is harrison')

print(cos_sim(BGE_query_q, BGE_query_a))
print(cos_sim(BGE_query_q, embeddings[1]))
print(cos_sim(BGE_query_q, embeddings[3]))
```

# ko-sbert-nli

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings

model_name = 'jhgan/ko-sbert-nli'
model_kwargs = {'device':'cpu'}
encode_kwargs = {'normalize_embeddings':True}

ko = HuggingFaceBgeEmbeddings(
        model_name = model_name,
        model_kwargs = model_kwargs,
        encode_kwargs = encode_kwargs
)
```

```python
sentence = [
    '안녕하세요',
    '제 이름은 홍길동 입니다.',
    '이름이 무엇인가요?',
    '렝체인은 유용합니다.',
    '홍길동 아버지의 이름은 홍상직 입니다.'
]       # 임베딩 sentence 문구

ko_embeddings = ko.embed_documents(sentence)
```

```python
q = '홍길동은 아버지를 아버지라 부르지 못했습니다. '
a = '홍길동의 아버지는 엄했습니다.'

ko_query_q = ko.embed_query(q)
ko_query_a = ko.embed_query(a)

print(cos_sim(ko_query_q, ko_query_a))          # 두 문장간의 코사인 유사도 검증
print(cos_sim(ko_query_q, ko_embeddings[1]))    # 두 문장간의 코사인 유사도 검증 (임베딩1)
print(cos_sim(ko_query_q, ko_embeddings[3]))    # 두 문장간의 코사인 유사도 검증 (임베딩3)
print(cos_sim(ko_query_q, ko_embeddings[4]))    # 두 문장간의 코사인 유사도 검증 (임베딩4)
```
